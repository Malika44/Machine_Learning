```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2.1,3.9,6.0,7.9,10.2])
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
print(f"Intercept: {model.intercept_}")
print(f"Slope:{model.coef_}")
```

```python
X_new = np.array([[6],[7],[8]])
```

```python
y_pred = model.predict(X_new)
for i, pred in enumerate(y_pred):
    print(f"x = {X_new[i][0]} predction is  ŷ = {pred:.2f}")
```

```python
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='green', label='Fitted Line')
plt.scatter(X_new, y_pred, color='red', label='New Predictions')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.title('OLS Regression & Predictions')
plt.grid(True)
plt.show()
```

Create some noise

```python
X1 = np.array([[1], [2], [3], [4], [5]])    # shape: (5, 1)
y2 = 2 * X1.flatten() + np.random.normal(0, 0.5, size=X1.shape[0])
```

```python
noise = np.random.normal(loc=0, scale=0.5, size=y1.shape) 
y2 = y1 + noise
```

```python
model1 = LinearRegression()
model1.fit(X, y2)
```

```python
X3 = np.array([[6], [7], [8]])
y_pred1 = model1.predict(X3)
```

```python
print(f"Intercept: {model1.intercept_}")
print(f"Slope:{model1.coef_}")
```

```python
for i, pred in enumerate(y_pred1):
    print(f"x = {X3[i][0]} prediction is ŷ = {pred:.2f}")
```

```python
print("X1 shape:", X1.shape)
print("y2 shape:", y2.shape)
```

```python
plt.scatter(X1, y2, color='blue', label='Noisy Data')
plt.plot(X1, model1.predict(X1), color='green', label='best Fit line')
plt.scatter(X3, y_pred1, color='red', label='Predictions')
plt.plot(X1, y2, color='gray', linestyle='--', label='True Line (No Noise)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('OLS with Noisy Data')
plt.legend()
plt.grid(True)
plt.show()
```
