```python
import tensorflow as tf
import numpy as np
```

```python
X = tf.random.uniform((100, 2))
y = (X+5)*0.3
```

```python
X.shape, y.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(name= "InputLayer", shape =[2]),
    tf.keras.layers.Dense(100, name= "Hidden_layer_1"),
    tf.keras.layers.Dense(200, name ="Hidden_layer_2"),
    tf.keras.layers.Dense(10, name = "Hidden_layer_3"),
    tf.keras.layers.Dense(2, name = "Output_layer")

])
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_layer_names=True, show_shapes=True)
```

```python
model.compile(loss = "MAE", optimizer= "Adam", metrics= ["MSE"])
```

```python
model.fit(X, y, epochs= 100)
```
